```python
# 라이브러리 로드

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # 전체를 다 가져오기엔 너무 크고 무거워
import numpy as np
```

```python
## matplotlib을 이용한 그래프의 시각화

# 라인 그래프 (플롯 그래프)
    # y 축의 데이터만 존재하는 경우
    # x 축은 0부터 시작해서 1씩 증가하는 형태로 생성됨

ylist = [10, 5, 8, 16]
plt.plot(ylist)

# show() == 표의 print()

plt.show()
```

```python
    # x, y 축의 데이터가 모두 존재하는 경우

x_list = [1, 3, 6, 10]
y_list = [1, 2, 3, 4]

plt.plot(x_list, y_list)
plt.show()
```

```python
    # line의 색상과 종류를 변경

plt.plot(x_list, y_list, 'r')
plt.show()
```

```python
plt.plot(x_list, y_list, '#FF0077')
plt.show()
```

```python
plt.plot(x_list, y_list, '--')

# x 축에 라벨링
plt.xlabel('X')

 # y 축에 라벨링
plt.ylabel('Y')

 # 그래프의 이름 표시

plt.title('Graph') # 한글 쓰면 넴몸넴몸 -> 기본 matplotlib 폰트가 한글을 지원하지 않음 

# 그래프에 범례 추가

plt.legend(['a', 'b'])

plt.show()
```

```python
df = pd.read_csv('../../csv/drinks.csv')
```

```python
df.head()
```

```python
# 맥주 소비량, 위스키(스피릿) 소비량, 와인의 소비량, 순수 알코올양의 상관관계를 확인
# 상관 계수 == -1 ~ 1
    # -1 == 하나의 데이터가 오르면 나머지 하나의 데이터는 내려간다, 반비례
    # 1 == 하나의 데이터가 오르면 나머지 하나의 데이터도 같이 오른다, 정비례
    # 0 == 두 개의 데이터가 무작위이다

# DataFrame 안에 있는 corr() 함수 호출 co realation ?

corr = df[ ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].corr('pearson') # 피어슨 
```

```python
# seaborn 라이브러리를 이용하여 히트맵 시각화

cols = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    corr.values,
    cbar= True, # 사이드바 표시
    annot = True, # 히트맵에 숫자를 표시
    square = True, # 정사각형 표시
    fmt = '0.3f', # 숫자의 소수점 자리수
    annot_kws= {'size' : 20}, # 숫자의 크기
    xticklabels= cols,
    yticklabels= cols
    )

plt.show
```

```python
sns.pairplot(
    df.loc[:, 'beer_servings': 'total_litres_of_pure_alcohol'],
    height= 2.5,
    kind = 'reg'# 보조선
)

plt.show()
```
