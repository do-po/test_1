```python
from wordcloud import WordCloud
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# txt 파일을 로드 
text = open("hiphop.txt", encoding="euc-kr").read()
```

```python
text
```

```python
okt = Okt()
```

```python
## Okt를 이용해서 문자열의 형태를 분석 
nouns = okt.nouns(text)
```

```python
# 글자수가 2개 이상인 문자만 필터
words = [ n for n in nouns if len(n) > 1 ]
```

```python
# 위와 같은 방법 (위는 한 줄에 쓰는 방법이고, 아래는 풀어 사용하는 방법)

words = []

for n in nouns:
    if len(n) > 1:
        words.append(n)
    else:
        print('배고프다')
```

```python
# 이것도 위와 같은 방법 (python 내장 map 사용)

list(
    map(
    lambda x : x if len(x) >= 2 else (print('배고프다')),
    nouns
)
)
```

```python
# 글자들의 빈도수를 출력 
cnt = Counter(words)
```

```python
# 아마 이 방법이 아래의 try 방법 보다 더 빠를거야 -> if는 조건에 맞는 것 2 중 1만 실행이 되지만 try는 2회 실행의 가능성이 있기에 try가 미세하게 더 느릴 듯

cnt_test = {}

for n in words:
    if n in cnt_test:
        cnt_test[n] += 1
    else:
        cnt_test[n] = 1

cnt_test

sorted(cnt_test, )
```

```python
cnt_test2 = {}

for n in words:
    try:
        cnt_test2[n] +=1
    except:
        cnt_test2[n] = 1

cnt_test2
```

```python
img = Image.open('pngwing.png')
img_array = np.array(img)

wc = WordCloud(
    font_path='malgun',
    width = 400, 
    height = 400, 
    scale = 2.0, # 최소 크기와 최대 크기 사이의 비율, 지금은 최소 크기를 2배로 확장한 것
    max_font_size= 250, 
    max_words = 100,
    mask=img_array # 워드 클라우드 배열을 이미지 배열로 만들어 주는 것 (이미지 행렬로 좌표를 만들어 경계, 구분을 만들어 주는 것)
)
gen = wc.generate_from_frequencies(cnt)
plt.figure()
plt.axis('off')
plt.imshow(gen)
```
