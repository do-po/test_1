## 01. 데이터 전처리

#### 01-1 서울시 스타벅스 매장 목록, 인구, 사업체 데이터에 시군구명, 시군구 코드 추가

```python
# 서울시 스타벅스 매장 목록이 담긴 파일 불러오기
import pandas as pd

seoul_starbucks = pd.read_csv('./files/starbucks.csv')

seoul_starbucks.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
# 스타벅스 주소 정보에서 시군구명 추출
    # 주소를 통해 구를 추출
    # 추출된 구를 다른 데이터들과 조인할 매개체로 사용하지 않을까 ??

sgg_names = []

for address in seoul_starbucks['주소']:
    sgg = address.split()[1]
    sgg_names.append(sgg)

seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()
```

```python
seoul_starbucks.to_excel('./files/seoul_starbucks_list.xlsx', index= False)
```

## 01-2 스타벅스 분석 데이터 만들기

```python
seoul_sgg = pd.read_excel('./files/seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
pivot = seoul_starbucks.pivot_table(
    values= '매장명',
    index= '시군구명',
    aggfunc= 'count'
)
```

```python
seoul_starbucks_total = pd.merge(seoul_sgg, pivot, on= '시군구명')
```

```python
seoul_starbucks_total.rename(columns= {'매장명' : '스타벅스_매장수'}, inplace= True)
```

```python
seoul_starbucks_total
```

```python
# 서울시 시군구 목록 데이터에 서울시 시군구별 사업체 수 통계 데이터를 병합

seoul_sgg_biz = pd.read_excel('./files/sgg_biz.xlsx')

seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_biz, how= 'left', on= '시군구명')
```

```python
seoul_sgg.to_excel('./files/seoul_sgg_stat.xlsx', index= False)
```
