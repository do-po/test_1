```python
import pandas as pd
import folium
import json

# 서울시 스타벅스 매장 목록 데이터 불러오기
seoul_starbucks = pd.read_excel('./files/seoul_starbucks_list.xlsx')
seoul_starbucks.head()
```

```python
# folium을 이용한 지도 생성
starbucks_map = folium.Map(
    location=[37.573050, 126.979189],
    zoom_start=11
)
starbucks_map
```

```python
# 스타벅스 매장 위치를 나타내는 서클 마커 그리기
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx, '위도']
    lng = seoul_starbucks.loc[idx, '경도']
    
    folium.CircleMarker(
        location=[lat, lng],
        fill = True,
        fill_color='green',
        fill_opacity=1, # 투명도
        color='yellow',
        weight=1, # 색 채우겠다
        radius=3
    ).add_to(starbucks_map)

starbucks_map
```

```python
# 스타벅스 매장 위치를 나타내는 서클 마커 그리기

for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx, '위도']
    lng = seoul_starbucks.loc[idx, '경도']

    if seoul_starbucks.loc[idx, '매장타입'] == 'pin_general':
        colors = 'grey'
    elif seoul_starbucks.loc[idx, '매장타입'] == 'pin_reserve':
        colors = 'blue'
    elif seoul_starbucks.loc[idx, '매장타입'] == 'pin_generalDT':
        colors = 'red'
    elif seoul_starbucks.loc[idx, '매장타입'] == 'pin_generalWT':
        colors = 'green'
    else:
        colors = 'purple'

    folium.CircleMarker(
        location=[lat, lng],
        fill = True,

        fill_color=f'{colors}',
        fill_opacity=1, # 투명도
        color=f'{colors}',
        weight=1, # 색 채우겠다
        radius=3
    ).add_to(starbucks_map)

starbucks_map
```
