```python
from folium.map import Popup
import pandas as pd
import folium
```

```python
df = pd.read_excel(f'{path_Data_in}서울지역 대학교 위치.xlsx')
```

```python
df
```

```python
seoul_map = folium.Map(location = [37.55,126.98], zoom_start = 12)
```

```python
seoul_map
```

```python
df
```

```python
df.set_index('Unnamed: 0', inplace = True)
```

### IFrame popup

```python
# 대학교 위치정보를 Marker로 표시
for name, lat, lng in zip(df.index, df.위도, df.경도): # zip == record(행의 단위)를 불러올 수 있는 함수
    iframe = folium.IFrame(name, width = 230, height = 50) # popup 객체 속의 IFrame의 사이즈 변경 (popup 레이어 위의 객체 저장 레이어)
    popup = folium.Popup(iframe, max_width = 300) # 선택했을 때 뒷 배경이 클릭되지 않도록 배경지를 만들어 주는 레이어
    folium.Marker([lat, lng], popup=popup).add_to(seoul_map) # [위도, 경도], index로 마커 팝업

seoul_map.save(f'{path_Data_out}seoul.colleges.html')
seoul_map
```

### CircleMarker

```python
for name, lat, lng in zip(df.index, df.위도, df.경도): # zip == record(행의 단위)를 불러올 수 있는 함수
    folium.CircleMarker([lat, lng],
                        radious = 10, # 원의 반지름
                        color = 'brown', # 원의 둘레 색상
                        fill = True, # 채우기 활성화
                        fill_color = 'coral', # 원을 채우는 색
                        fill_opacity = 0.7, # 투명도
                        popup=name).add_to(seoul_map) # [위도, 경도], index로 마커 팝업

seoul_map.save(f'{path_Data_out}seoul.colleges2.html')
seoul_map
```

### 경기도 인구변화

```python
import json
```

```python
file_path = f'{path_Data_in}경기도인구데이터.xlsx'

df1 = pd.read_excel(file_path, index_col='구분', engine= 'openpyxl')
```

```python
df1.head()
```

```python
df1.columns = df1.columns.map(str)
```

```python
df1.columns
```

```python
# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기

geo_path = f'{path_Data_in}경기도행정구역경계.json'

try:
    geo_data = json.load(open(geo_path, encoding= 'utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding= 'utf-8-sig'))
```

```python
geo_data
```

```python
# 경기도 지도 만들기

g_map = folium.Map(location= [37.5502, 126.982], #tiles = 'Stamen Terrain' 으로 하면 지도가 안 나와 :> attr을 잘 넣어야 하나봐
                   zoom_start= 9)

# 출력할 연도 선택

year = '2017'

# Choropleth 클래스로 단계 구분도 표시하기

folium.Choropleth(geo_data= geo_data,
                  data= df1[year],
                  columns= [df1.index, df1[year]],
                  fill_color= 'YlOrRd', fill_opacity= 0.3, line_opacity= 0.2,
                  threshold_scale = [10000,10000, 300000, 500000, 700000],
                  key_on = 'feature.properties.name').add_to(g_map)

g_map.save(f'{path_Data_out}' + f'{year}.html')
```
