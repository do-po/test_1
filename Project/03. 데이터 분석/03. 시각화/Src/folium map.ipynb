```python
from folium.map import Popup
import pandas as pd
import folium
```

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx')
```

```python
df
```

```python
seoul_map = folium.Map(location = [37.55,126.98], zoom_start = 12)
```

```python
seoul_map
```

```python
df
```

```python
df.set_index('Unnamed: 0', inplace = True)
```

### IFrame popup

```python
# 대학교 위치정보를 Marker로 표시
for name, lat, lng in zip(df.index, df.위도, df.경도): # zip == record(행의 단위)를 불러올 수 있는 함수
    iframe = folium.IFrame(name, width = 230, height = 50) # popup 객체 속의 IFrame의 사이즈 변경 (popup 레이어 위의 객체 저장 레이어)
    popup = folium.Popup(iframe, max_width = 300) # 선택했을 때 뒷 배경이 클릭되지 않도록 배경지를 만들어 주는 레이어
    folium.Marker([lat, lng], popup=popup).add_to(seoul_map) # [위도, 경도], index로 마커 팝업

seoul_map.save('./seoul.colleges.html')
seoul_map
```

### CircleMarker

```python
for name, lat, lng in zip(df.index, df.위도, df.경도): # zip == record(행의 단위)를 불러올 수 있는 함수
    folium.CircleMarker([lat, lng],
                        radious = 10, # 원의 반지름
                        color = 'brown', # 원의 둘레 색상
                        fill = True, # 채우기 활성화
                        fill_color = 'coral', # 원을 채우는 색
                        fill_opacity = 0.7, # 투명도
                        popup=name).add_to(seoul_map) # [위도, 경도], index로 마커 팝업

seoul_map.save('./seoul.colleges2.html')
seoul_map
```
