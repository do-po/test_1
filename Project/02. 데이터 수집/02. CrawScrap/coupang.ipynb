```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

service = Service()


options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python

driver.get("https://www.coupang.com/np/categories/194276")

productList = driver.find_element(By.ID, 'productList')

lis = productList.find_elements(By.TAG_NAME, 'li')

prd_name = []
prd_price = []
prd_deli = []
prd_url = []
prd_foodname = []

for idx, li in enumerate(lis):
    try:
        prd_name.append(li.find_element(By.CLASS_NAME, 'name').text)
        prd_price.append(li.find_element(By.CLASS_NAME, 'price-value').text)
        prd_deli.append(li.find_element(By.CLASS_NAME, 'delivery').text)
        prd_url.append(li.find_element(By.CLASS_NAME, 'baby-product-link').get_attribute('href'))


        # image 저장

        e = li.find_element(By.TAG_NAME,'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()
        food_name = "./img/food{}.jpg".format(idx+1)

        with open(food_name, "wb") as f:
            f.write(img_res.content)
        prd_foodname.append(food_name)

        if idx >= 5:
           break

    except Exception:
        pass



final = pd.DataFrame({'이름' : prd_name, '가격' : prd_price, 'Delivery' : prd_deli,
                      'URL' : prd_url, 'foodname':prd_foodname})

# print(len(prd_name),len(prd_price), len(prd_deli), len(prd_url))

final.to_csv('쿠팡식품.csv', encoding='euc-kr')
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.e-himart.co.kr/app/display/showDisplayCategory?dispNo=1011030000")

productList = driver.find_element(By.ID, 'productList')

lis = productList.find_elements(By.TAG_NAME, 'li')
```

```python
prd_name = []
prd_price = []
prd_deli = []
prd_star = []
prd_url = []
prd_imgname = []
```

```python
for idx, li in enumerate(lis):
    try:

        # prd_name

        prd_name.append(li.find_element(By.CLASS_NAME, 'prdName').text)
        
        # prd_price

        tmp_price = li.find_elements(By.CLASS_NAME, 'discountPrice')
        
        if (len(tmp_price) == 2):
            prd_price.append(tmp_price[1].text)
        else:
            prd_price.append(tmp_price[0].text)

        #prd_deli

        try:
            tmp_prd_deli = li.find_elements(By.CLASS_NAME, 'prdFlag')

            if (len(tmp_prd_deli) <= 2):
                prd_deli.append(tmp_prd_deli[-1].text)
            
            else:
                prd_deli.append(tmp_prd_deli[0].text)

        except:
            prd_deli.append('-')


        # prd_star

        prd_star.append(li.find_element(By.CLASS_NAME, 'ratingPoint').text)

        # prd_url
        
        prd_url.append(li.find_element(By.TAG_NAME, 'a').get_attribute('href'))
        
        # prd_imgname
        # image 저장

        e = li.find_element(By.TAG_NAME,'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()
        image_name = "./img/laundry{}.jpg".format(idx+1)

        with open(image_name, "wb") as f:
            f.write(img_res.content)
        prd_imgname.append(image_name)

        if idx >= 5:
           break

    except Exception:
        pass



final = pd.DataFrame({'이름' : prd_name, '가격' : prd_price, 'Delivery' : prd_deli,
                      '별점' : prd_star, 'URL' : prd_url})

# print(len(prd_name),len(prd_price), len(prd_deli), len(prd_url))

# final.to_csv('쿠팡식품.csv', encoding='euc-kr')
```
